1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
df = pd.read_csv("previsao_de_renda.csv").drop('Unnamed: 0', axis=1).drop_duplicates()
df['data_ref'] = pd.to_datetime(df['data_ref'])
df['ano_ref'] = df['data_ref'].dt.year
df.head()
````

````python
def dist_ano_plot(df, var):
    tab = pd.crosstab(df['ano_ref'], df[var])
    tab = tab.div(tab.sum(axis=1), axis=0)
    return tab.plot.bar()
````

````python
def media_ano_plot(df, var):
    return sns.pointplot(data=df, x='ano_ref', y=var)
````

````python
dist_ano_plot(df, 'sexo')
````
